```python
import pandas as pd

df = pd.read_csv("data/AI_Impact_on_Jobs_2030.csv")
df.head()
```

```python
df.info()
df.columns
df.describe()
```

```python
def categorize_risk(prob):
    if prob < 0.3:
        return 'Low Risk'
    elif prob < 0.7:
        return 'Medium Risk'
    else:
        return 'High Risk'

df['Risk_Category'] = df["Automation_Probability_2030"].apply(categorize_risk)
```

```python
df["Risk_Category"].value_counts()
```

```python
import matplotlib.pyplot as plt

df["Risk_Category"].value_counts().plot(kind='bar')
plt.title("Job Risk Categories in 2030")
plt.ylabel("Number of Jobs")
plt.show()
```

```python
# First make sure Risk_Category exists
df["Risk_Category"] = pd.cut(
    df["Automation_Probability_2030"],
    bins=[0, 0.3, 0.6, 1],
    labels=["Low Risk", "Medium Risk", "High Risk"]
)

# Calculate average salary per risk category
salary_comparison = df.groupby("Risk_Category")["Average_Salary"].mean()

salary_comparison
```

```python
import matplotlib.pyplot as plt

salary_comparison.plot(kind="bar")
plt.title("Average Salary by Automation Risk Category (2030)")
plt.ylabel("Average Salary")
plt.xlabel("Risk Category")
plt.show()
```

```python
import matplotlib.pyplot as plt

plt.figure()
for category in ["Low Risk", "Medium Risk", "High Risk"]:
    subset = df[df["Risk_Category"] == category]
    plt.hist(subset["Average_Salary"], alpha=0.5, bins=30, label=category)

plt.xlabel("Average Salary")
plt.ylabel("Frequency")
plt.title("Salary Distribution by Risk Category")
plt.legend()
plt.show()

```

```python
low = salary_comparison["Low Risk"]
high = salary_comparison["High Risk"]

percentage_difference = ((low - high) / high) * 100
percentage_difference
```
